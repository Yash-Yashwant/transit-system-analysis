```python
import glob
import pandas as pd
```

```python
# Use glob to get all the CSVs.
files = glob.glob("../../raw_data/Ottawa_Transit_Data/*.csv")
```

```python
# Initialize storage file.
data = None
# Loop over individual data files.
for file in files:
    # Use first file to initialize.
    if data is None:
        data = pd.read_csv(file)
        # Create column to join on.
        data['time_stamp'] = [m + ' ' + str(y) for m,y in zip(data['Month'], data["Year"])]
    else:
        temp = pd.read_csv(file)
        # Create column to join on.
        temp['time_stamp'] = [m + ' ' + str(y) for m,y in zip(temp['Month'], temp["Year"])]
        # Remove duplicate columns.
        temp = temp.drop(columns=['Month', 'Year'])
        # Join data frames.
        data = data.join(temp.set_index('time_stamp'), on='time_stamp')
# Drop now redundant join column.
data = data.drop(columns=['time_stamp'])
data.head(5)
```

```python
# Save combined result.
data.to_csv('../../raw_data/Ottawa_Transit_Data/ottawa_all.csv', sep=',')
```
